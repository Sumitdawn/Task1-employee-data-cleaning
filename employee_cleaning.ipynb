```python
import pandas as pd
import numpy as np
df = pd.read_csv("employee_raw_dataset.csv")
df.head()              
df.info()              
df.describe(include='all') 
```

```python
df.isnull().sum() 
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df['Gender'].fillna('Unknown', inplace=True)
df['Department'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Joining Date'].fillna('01-01-2000', inplace=True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df['Gender'] = df['Gender'].fillna('Unknown')
df['Department'] = df['Department'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Joining Date'] = df['Joining Date'].fillna('01-01-2000')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Gender'] = df['Gender'].str.lower().map({
    'male': 'Male', 'm': 'Male',
    'female': 'Female', 'f': 'Female',
    'other': 'Other',
    'unknown': 'Unknown'
})
df['Country'] = df['Country'].replace({
    'United States': 'USA',
    'U.S.': 'USA',
    'IND': 'India',
    'United Kingdom': 'UK'
})
```

```python
df['Joining Date'] = pd.to_datetime(df['Joining Date'], errors='coerce')
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['age'] = df['age'].astype(int)
df['salary'] = df['salary'].astype(float)
```

```python
df.to_csv("employee_cleaned_dataset.csv", index=False)
```
